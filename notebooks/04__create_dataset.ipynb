```python
print("=" * 70)
print("CR√âATION DU DATASET PRINCIPAL")
print("=" * 70)

# Param√®tres du dataset
N_VALID = 5000      # 5000 √©tats valides
N_INVALID = 5000    # 5000 √©tats invalides
DIM = 4             # Dimension 4 (compromis complexit√©/interpr√©tabilit√©)
SEED = 42           # Reproductibilit√©

# Cr√©e le dataset
df = create_dataset(
    n_valid=N_VALID,
    n_invalid=N_INVALID,
    dim=DIM,
    valid_strategy="random",      # √âtats valides: strat√©gie random
    invalid_strategy="mixed",     # √âtats invalides: strat√©gie mixed (diversit√© max)
    invalid_kwargs={
        "extreme_prob": 0.10,     # 10% de cas extr√™mes dans les invalides
        "scale_range": (0.1, 2.5) # Large plage pour scaling
    },
    seed=SEED,
    shuffle=True  # M√©lange pour √©viter biais d'ordre
)

print("\n Dataset cr√©√© avec succ√®s!")
```

```python
print("=" * 70)
print("EXPLORATION DU DATASET")
print("=" * 70)

# Affiche les infos structur√©es
print_dataset_info(df)

# Affiche les premi√®res lignes
print("\nüìã Premi√®res lignes:")
print(df.head(10))

# Types de colonnes
print("\nüî¢ Types de donn√©es:")
print(df.dtypes)

# V√©rifie valeurs manquantes
print("\nüîç Valeurs manquantes:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("    Aucune valeur manquante")
else:
    print(missing[missing > 0])

# Statistiques descriptives
print("\n Statistiques descriptives (features num√©riques):")
print(df.describe())
```
