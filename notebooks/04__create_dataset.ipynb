```python
"""
Notebook: 04_create_dataset.ipynb
Objectif: Créer le dataset complet pour entraînement ML
Auteur: Mklzenin
Date: 2024-11-12
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# Import du module
project_root = Path.cwd().parent
sys.path.insert(0, str(project_root / "src"))

from data_generation import (
    create_dataset,
    save_dataset,
    load_dataset,
    print_dataset_info,
    get_dataset_info
)

print("Module importé avec succès!")
print(f"Projet: {project_root}")
```

```python
print("=" * 70)
print("CRÉATION DU DATASET PRINCIPAL")
print("=" * 70)

# Paramètres du dataset
N_VALID = 5000      # 5000 états valides
N_INVALID = 5000    # 5000 états invalides
DIM = 4             # Dimension 4 (compromis complexité/interprétabilité)
SEED = 42           # Reproductibilité

# Crée le dataset
df = create_dataset(
    n_valid=N_VALID,
    n_invalid=N_INVALID,
    dim=DIM,
    valid_strategy="random",      # États valides: stratégie random
    invalid_strategy="mixed",     # États invalides: stratégie mixed (diversité max)
    invalid_kwargs={
        "extreme_prob": 0.10,     # 10% de cas extrêmes dans les invalides
        "scale_range": (0.1, 2.5) # Large plage pour scaling
    },
    seed=SEED,
    shuffle=True  # Mélange pour éviter biais d'ordre
)

print("\n Dataset créé avec succès!")
```
