# 04 - Main dataset creation

This notebook is the **executable documentation of the shipped dataset**: the
parameters below are exactly those of the versioned CSV, so re-running it must
reproduce `data/processed/quantum_states_10000.csv` byte for byte.

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if not (ROOT / "src").exists():
    ROOT = ROOT.parent
if str(ROOT / "src") not in sys.path:
    sys.path.insert(0, str(ROOT / "src"))
import numpy as np
import pandas as pd

from qsv.paths import DATA_PROCESSED, MAIN_DATASET
from qsv.data_generation import create_dataset, save_dataset

print("Modules imported, project root:", ROOT.name)
```

```python
# Dataset parameters - EXACTLY those of the versioned dataset
# (documented in data/README.md: create_dataset(5000, 5000, dim=4, seed=42)
# with the default strategies).
N_VALID = 5000
N_INVALID = 5000
DIM = 4
SEED = 42

df = create_dataset(
    n_valid=N_VALID,
    n_invalid=N_INVALID,
    dim=DIM,
    seed=SEED,
    shuffle=True,
)

print(f"Dataset created: {df.shape}")
df.head()
```

```python
# Save anchored on the project root (qsv/paths.py) - no cwd-dependent
# duplicate can ever be created again (audit A1).
filepath = save_dataset(df, filename=MAIN_DATASET.name, data_dir=str(DATA_PROCESSED))
print(f"Dataset saved: {filepath}")
print(f"Size: {filepath.stat().st_size / 1024:.2f} KB")
```
