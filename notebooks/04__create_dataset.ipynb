```python
# Bootstrap : rendre src/ importable et ancrer les chemins sur la racine projet
import sys
from pathlib import Path

ROOT = Path.cwd()
if not (ROOT / "src").exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.paths import DATA_PROCESSED, MAIN_DATASET
from src.data_generation import (
    create_dataset,
    save_dataset,
    load_dataset,
    print_dataset_info,
)

print("Modules importés, racine projet :", ROOT.name)
```

```python
print("=" * 70)
print("CRÉATION DU DATASET PRINCIPAL")
print("=" * 70)

# Paramètres du dataset — EXACTEMENT ceux du dataset versionné
# (documentés dans data/README.md : create_dataset(5000, 5000, dim=4, seed=42)
# avec les stratégies par défaut). Ce notebook est la documentation
# exécutable du dataset : le ré-exécuter doit reproduire le CSV à
# l'identique, octet pour octet.
N_VALID = 5000
N_INVALID = 5000
DIM = 4
SEED = 42

df = create_dataset(
    n_valid=N_VALID,
    n_invalid=N_INVALID,
    dim=DIM,
    seed=SEED,
    shuffle=True,
)

print(f"\nDataset créé : {df.shape}")
df.head()
```

```python
# Sauvegarde ancrée sur la racine du projet (src/paths.py) — plus jamais
# de duplicata notebooks/data/ créé par un mauvais cwd (audit A1).
filepath = save_dataset(df, filename=MAIN_DATASET.name, data_dir=str(DATA_PROCESSED))
print(f"Dataset sauvegardé : {filepath}")
print(f"Taille : {filepath.stat().st_size / 1024:.2f} KB")
```

```python
print("=" * 70)
print("EXPLORATION DU DATASET")
print("=" * 70)

print_dataset_info(df)

# Affiche les premières lignes
print("\n Premières lignes:")
print(df.head(10))

# Types de colonnes
print("\n Types de données:")
print(df.dtypes)

# Vérifie valeurs manquantes
print("\n Valeurs manquantes:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("    Aucune valeur manquante")
else:
    print(missing[missing > 0])

# Statistiques descriptives
print("\n Statistiques descriptives (features numériques):")
print(df.describe())
```
